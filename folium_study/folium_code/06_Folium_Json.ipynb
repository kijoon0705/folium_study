## 2. GeoJson

```python
m = folium.Map(location= [36, 128], zoom_start= 7, tiles='OpenStreetMap')

geo_path = '/home/seong/workspace/folium_study/data/korea.json'
geo_str = json.load(open(geo_path, mode='rt', encoding='utf-8'))

m.choropleth(geo_data=geo_str)
m.save('korea_json.html')

m
```
